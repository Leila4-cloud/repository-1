```python
import gspread
from google.auth import default

creds, _ = default()
gc = gspread.authorize(creds)


sheet = gc.open("covariance").sheet1
```

```python
values = sheet.get_all_values()
print("Sheet content:")
for row in values:
    print(row)
```

```python
values
```

```python
import pandas as pd

column_headers = values[0]
data_rows = values[1:]

df_returns = pd.DataFrame(data_rows, columns=column_headers)
print("DataFrame 'df_returns' created successfully.")
```

```python
columns_to_convert = [col for col in df_returns.columns if col != 'DATES']

numerical_returns = df_returns[columns_to_convert].copy()

for col in columns_to_convert:
    numerical_returns[col] = numerical_returns[col].str.replace('%', '', regex=False).astype(float) / 100

print("Percentage columns cleaned and converted to numerical format.")
print(numerical_returns.head())
```

```python
covariance_matrix = numerical_returns.cov()
print("Covariance matrix calculated successfully.")
print(covariance_matrix.head())
```
